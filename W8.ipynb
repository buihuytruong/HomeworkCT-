[Cylinder]
Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và chiều cao
(kiểu số thực) với hai phương thức:tính diện tích bề mặt (diện
tích toàn phần) và thể tích của hình trụ
Đầu vào
Đâu vào từ bàn phím gồm một dòng duy nhất chứa bán kính và
chiều cao của hình trụ, cách nhau bởi một dấu cách.
Đầu ra
In ra màn hình diện tích bề mặt và thể tích của hình trụ (sử
dụng lại những hàm vừa viết, π=3.14).
(Kết quả làm tròn đến 2 chữ số sau dấu phẩy). 

```python
class Cylinder():
    def __init__(self , rr , hh):
        self.r = rr
        self.h = hh
    
    def getArea(self) -> float:
        return (3.14*self.r**2)*2 + (3.14*2*self.r*self.h)
    def getV(self) -> float:
        return self.r**2*3.14* self.h
    
    def __str__(self):
        return f'({self.r}),({self.h})'
    def __repr__(self):
        return f'({self.r}),({self.h})'
m,n = map(float,input().split())
cy1 = Cylinder(m,n)
print(cy1.getArea())
print(cy1.getV())
```

[Date]
Xây dựng lớp Date biểu diễn ngày tháng năm với hai phương
thức kiểm tra xem một ngày tháng nào đó có hợp lệ hay không
và phương thức tính ngày kế tiếp của ngày tháng năm nào đó
nếu nó là hợp lệ
Nhập vào một ngày tháng năm, (ngày tháng năm biểu diễn
dưới dạng dd/mm/yyyy). Nếu ngày tháng năm không tồn tại, in
ra dòng chữ INVALID. Nếu là hợp lệ in ra ngày kế tiếp của nó.
Đầu vào
Đầu vào từ bàn phim chứa dữ liệu ngày, tháng, năm dưới dạng
dd/mm/yyyy
.
Đầu ra
In ra màn hình ngày kế tiếp của ngày vừa nhập dưới dạng
dd/mm/yyyy
. Nếu ngày tháng không tồn tại, in ra INVALID

```python
class Date():
    def __init__(self,d,m,y):
        self.d = d
        self.m = m
        self.y = y
    def __str__(self):
        return f'{self.d:02d}/{self.m:02d}/{self.y:02d}'
    def check(self):
        maxDay = -1
        match self.m:
            case 1|3|5|7|8|10|12:
                maxDay = 31
            case 2:
                if self.y % 4 == 0 and self.y%100 != 0 or self.y % 400 == 0:
                    maxDay = 29
                else:
                    maxDay = 28
            case 4|6|9|11:
                maxDay = 30
        if 1 <= self.d <= maxDay and self.y > 0:
            return True
        else:
            return False
    def getmaxDay(self):
        match self.m:
            case 1|3|5|7|8|10|12:
                return 31
            case 2:
                if self.y % 4 == 0 and self.y%100 != 0 or self.y % 400 == 0:
                    return 29
                else:
                    return 28
            case 4|6|9|11:
                return 30
    def nextDay(self) -> 'Date':
        maxDay = self.getmaxDay()
        
        next_d = self.d + 1
        next_m = self.m
        next_y = self.y
        
        if next_d > maxDay:
            next_d = 1
            next_m += 1
            
            if next_m > 12:
                next_m = 1
                next_y += 1
                
        return Date(next_d, next_m, next_y)
        
        
d , m , y = map(int,input().split())
date1 = Date(d,m,y)
if date1.check() == True:
    print(date1)
    print(date1.nextDay())
else:
    print("False")
```

[Calculator] 
Xây dựng lớp Calculator (máy tính bỏ túi) với các phương thức 
để thực hiện các phép toán 1. add (cộng), 2. subtract(trừ), 3. 
product (nhân) và 4. divide(chia), 5. power(hàm mũ), 6. mod 
(lấy phần dư) và phương thức 7. set_numbers(cập nhật 2 số 
mới) nếu người dùng muốn tính toán trên 2 số khác. 
Khởi tạo lớp là hai thuộc tính gồm 2 số cần tính, Mỗi phương 
thức thực hiện một phép tính và trả về kết quả tương ứng 
Tạo menu và cho người dùng tính toán tới khi gõ "yes", "YES" 
hoặc "Yes" cho câu hỏi "Bạn có muốn thoát?" mới dừng.

```python
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2

    def product(self):
        return self.num1 * self.num2

    def divide(self):
        if self.num2 == 0:
            return "Lỗi: Không thể chia cho 0"
        return self.num1 / self.num2

    def power(self):
        return self.num1 ** self.num2

    def mod(self):
        if self.num2 == 0:
            return "Lỗi: Không thể chia lấy dư cho 0"
        return self.num1 % self.num2

    def set_numbers(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        print(f"Đã cập nhật 2 số mới: {self.num1} và {self.num2}")

a, b = map(float,input().split())
calc = Calculator(a, b)
while True:
    print("\n--- MENU MÁY TÍNH BỎ TÚI ---")
    print("1. Add (Cộng)")
    print("2. Subtract (Trừ)")
    print("3. Product (Nhân)")
    print("4. Divide (Chia)")
    print("5. Power (Hàm mũ)")
    print("6. Mod (Lấy phần dư)")
    print("7. Set numbers (Cập nhật 2 số mới)")
        
    choice = input("Chọn phép toán (1-7): ")

    if choice == '1':
        print(f"Kết quả: {calc.add()}")
    elif choice == '2':
        print(f"Kết quả: {calc.subtract()}")
    elif choice == '3':
        print(f"Kết quả: {calc.product()}")
    elif choice == '4':
        print(f"Kết quả: {calc.divide()}")
    elif choice == '5':
        print(f"Kết quả: {calc.power()}")
    elif choice == '6':
        print(f"Kết quả: {calc.mod()}")
    elif choice == '7':
        new_a = float(input("Nhập số thứ nhất mới: "))
        new_b = float(input("Nhập số thứ hai mới: "))
        calc.set_numbers(new_a, new_b)
    else:
        print("Lựa chọn không hợp lệ. Vui lòng chọn từ 1 đến 7.")
        print("Bạn có muốn thoát? ")
        thoat = input("\nBạn có muốn thoát? ")
        if thoat in ["yes", "YES", "Yes"]:
            print("Đã thoát chương trình.")
            break
```

[ShoppingCart] 
Xây dựng lớp ShoppingCart (giỏ hàng) để quản lý các sản 
phẩm của khách. Dùng danh sách để lưu các sản phẩm, mỗi 
sản phẩm là một từ điển có name (tên) và price(giá). 
Với các phương thức sau: 
1. Khởi tạo với danh sách rỗng 
2. Thêm vào giỏ hàng một sản phẩm mới 
3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên để xóa) 
4. Kiểm tra xem giỏ hàng có rỗng không 
5. Tính tổng tiền giỏ hàng 
6. Hiển thị toàn bộ sản phẩm trong giỏ 
7. Xóa toàn bộ giỏ hàng 
Tạo menu cho người dùng nhập lựa chọn truy vấn từ 2 tới 7. 
Chương trình dừng khi người dùng nhập "yes" cho câu hỏi "Có 
muốn thoát khỏi chương trình không?"

```python
class ShoppingCart():
    def __init__(self):
        self.p = []

    def add_products(self, name, price):
        product = {"name": name, "price": price}
        self.p.append(product)
        print(f'Da them {name} vao gio hang vs gia {price}')
    
    def remove_products(self, name):
        for i in self.p:
            if i["name"] == name:
                self.p.remove(i)
                print(f"Da xoa {name} khoi gio hang")
                return
        print(f"K tim thay {name} trong gio hang")
    
    def check(self):
        return len(self.p) == 0 
    
    def cal_products(self):
        return sum(item["price"] for item in self.p) 
    
    def display_products(self):
        if self.check() == True:
            print("Gio hang trong")
        else:
            print("Cac san pham:")
            for i, j in enumerate(self.p, 1):
                print(f"{i} . {j['name']} - {j['price']}")

    def clear_products(self):
        self.p.clear()
        print("Da don sach gio hang")

a = ShoppingCart()
while True:
    print("\n--- Menu Gio Hang ---")
    print("2. Thêm sản phẩm mới")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng rỗng")
    print("5. Tính tổng tiền giỏ hàng")
    print("6. Hiển thị toàn bộ sản phẩm")
    print("7. Xóa toàn bộ giỏ hàng")

    choice = input("Chọn chức năng: ")
    if choice == "2":
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá sản phẩm: "))
        a.add_products(name, price)  
        
    elif choice == '3':
        name = input("Nhập tên sản phẩm cần xóa: ")
        a.remove_products(name)
        
    elif choice == '4':
        if a.check() == True:
            print("Giỏ hàng đang TRỐNG.")
        else:
            print("Giỏ hàng ĐANG CÓ sản phẩm.")
            
    elif choice == '5':
        total = a.cal_products()
        print(f"Tổng tiền của giỏ hàng hiện tại là: {total}")
        
    elif choice == '6':
        a.display_products()
        
    elif choice == '7':
        a.clear_products()
        
    else:
        print("Lựa chọn không hợp lệ, vui lòng chọn từ 2 đến 7.")
        
    thoat = input("\nCó muốn thoát khỏi chương trình không? (yes để thoát): ")
    if thoat == "yes":
        print("Đã thoát chương trình.")
        break
```

Xây dựng lớp Rectangle(w, h) với các phương thức: 
area() trả về diện tích. 
perimeter() trả về chu vi. 
scale(k) phóng to hình chữ nhật theo hệ số k (nhân cả w và h), 
không trả về gì. 
Input 
w h k (số thực dương) 
Output 
area perimeter sau khi phóng to (làm tròn 2 chữ số thập phân)

```python
class Rectangle():
    def __init__(self, w, h):
        self.x = w
        self.y = h
    def area(self):
        return self.x * self.y
    def perimeter(self):
        return 2*(self.x +self.y)
    def scale(self,k):
        self.x = self.x * k
        self.y = self.y * k
w , h , k = map(float,input().split())
re1 = Rectangle(w,h)
re1.scale(k)
print(re1.area(),re1.perimeter())
        
```

Yêu cầu 
Lớp Fraction(num, den) (tử, mẫu nguyên, mẫu ≠ 0) với: 
Rút gọn phân số trong __init__ (quy ước mẫu > 0). 
Các phương thức add/sub/mul/div trả về phân số đã rút gọn. 
__str__() in dạng a/b. 
Input 
a b op c d (tương ứng a/b (op) c/d, op thuộc {+,-,*,/}) 
Output 
Phân số kết quả đã rút gọn

```python
class Fraction():
    def __init__(self,num,den):
        if den == 0:
            raise ValueError("Mau phai khac 0")
        
        if den < 0:
            num , den = -num , -den
        def UCLN(a,b):
            a , b  = abs(a) , abs(b)
            while b != 0:
                a , b = b , a%b
            return a
        g = UCLN(num,den)
        self.num = num // g
        self.den = den // g
    def add(self, other):
        return Fraction(self.num * other.den + other.num * self.den, self.den * other.den)

    def sub(self, other):
        return Fraction(self.num * other.den - other.num * self.den, self.den * other.den)

    def mul(self, other):
        return Fraction(self.num * other.num, self.den * other.den)

    def div(self, other):
        return Fraction(self.num * other.den, self.den * other.num)

    def __str__(self):
        return f"{self.num}/{self.den}"
a , b , op , c , d = input().split()
a = int(a)
b = int(b)
c = int(c)
d = int(d)
f1 = Fraction(a,b)
f2 = Fraction(c,d)
if op == "+":
    print(f1.add(f2))
if op == "-":
    print(f1.sub(f2))
if op == "*":
    print(f1.mul(f2))
if op == "/":
    print(f1.div(f2))
```

Yêu cầu 
Lớp BankAccount(owner, balance) với: 
deposit(x), withdraw(x); nếu rút quá số dư thì bỏ qua giao dịch.
__str__() in owner: balance. 
Input 
Dòng 1: owner initial_balance 
Dòng 2: n (số lệnh) 
Tiếp theo n dòng, mỗi dòng: DEPOSIT x hoặc WITHDRAW x. 
Output 
Số dư cuối cùng (in số thực, 2 chữ số).

```python
class BankAccount():
    def __init__(self,owner,balance):
        self.o = owner
        self.b = float(balance)
    
    def deposit(self,amount):
        if amount > 0:
            self.b += amount
    def withdraw(self,amount):
        if amount > 0 and amount <= self.b:
            self.b -= amount
    def __str__(self):
        return f'{self.o}:{self.b}:.2f'

bankacc1 = input().split()
owner , balance = bankacc1[0] , float(bankacc1[1])
acc1 = BankAccount(owner,balance)
n = int(input())
for i in range(n):
    lenh = input().split()
    action = lenh[0]
    amount = float(lenh[1])

    if action =='deposit':
        acc1.deposit(amount)
    elif action == 'withdraw':
        acc1.withdraw(amount)
print(f"{acc1.b:.2f}")
```

Lớp Student(name) với: 
add_score(subject, score) 
avg() trả về điểm trung bình. 
rank() trả về "Excellent" nếu avg ≥ 8, "Good" nếu 6.5–<8, 
"Average" nếu 5–<6.5, ngược lại "Poor". 
Input 
Dòng 1: name 
Dòng 2: n 
n dòng tiếp: subject score (score là số thực 0–10) 
Output 
name avg rank (avg 2 chữ số).

```python
class Student():
    def __init__(self,name):
        self.name = name
        self.score = []

    def add_score(self,subject,score):
        return self.score.append(score)
    def avg(self):
        return sum(self.score)/len(self.score)
    def rate(self):
        if self.avg() >= 8:
            return "Execellent"
        elif self.avg() >= 6.5: 
            return "Good"
        elif self.avg() >= 5:
            return "Avg"
        elif self.avg() < 5:
            return "Poor"
    def __str__(self):
        return f'{self.name} {self.avg()} {self.rate()}'
name = input()
n = int(input())
student = Student(name)
for i in range(n):
    data = input().split()
    sub = data[0]
    score = float(data[1])
    student.add_score(sub,score)
print(f'{student.name} {student.avg()} {student.rate()}')

```

Lớp Book(title, author, year). 
Lớp Library() với: 
add(book) 
count_by_author(name) 
find_by_year(y) trả về số sách xuất bản năm y. 
Input 
Dòng 1: n (số thao tác) 
n dòng: 
ADD title;author;year 
COUNT author (đếm theo tác giả) 
COUNTYEAR year (đếm theo năm) 
Output 
Mỗi lệnh COUNT/COUNTYEAR in một dòng kết quả.

```python
class Book():
    def __init__(self,title,author,year):
        self.t = title
        self.y = year
        self.a = author
    
class Library():
    def __init__(self):
        self.book = []
    
    def add(self,book):
        self.book.append(book)
    
    def count_by_author(self,name):
        return sum(1 for book in self.book if book.a == name)
    
    def count_by_year(self,year):
        return sum(1 for book in self.book if book.y == year)

n = int(input())
lib = Library()
for i in range(n):
    com = input().split()
    action = com[0]
    
    if action == "ADD":
        title , author ,year = com[1].split(";")
        lib.add(Book(title,author,int(year)))
    if action == "COUNT":
        author_name = com[1]
        print(lib.count_by_author(author_name))
    if action == "COUNTYEAR":
        year = int(com[1])
        print(lib.count_by_year(year))
        
```
