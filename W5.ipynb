```python
quad = []
for i in range(4):
    toado = input().split()

    x = float(toado[0])
    y = float(toado[1])

    quad.append((x,y))

pxy = input().split()
px = float(pxy[0])
py = float(pxy[1])
p = (px,py)

def stamgiac(a,b,c):
    x1 , y1 = a
    x2 , y2 = b
    x3 , y3 = c
    return abs((x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2))/2)

def check(quad , p):
    A , B , C, D = quad
    sabc = stamgiac(A,B,C)
    sacd = stamgiac(A,C,D)
    sabcd = sabc + sacd
    spab = stamgiac(p,A,B)
    spbc = stamgiac(p,B,C)
    spcd = stamgiac(p,C,D)
    spda = stamgiac(p,D,A)
    return sabcd - (spab + spbc +spcd + spda)

if check(quad, p) > 0:
    print("vien")
elif check(quad, p) == 0:
    print("trong")
else:
    print("ngoai")




```

```python
import math
diem = []
for i in range (3):
    toado = input().split()

    x = float(toado[0])
    y = float(toado[1])

    diem.append((x,y))

def degree(diem):
    A ,B ,C = diem
    x1 , y1 = A
    x2 , y2 = B
    x3 , y3 = C
    cosabc = ((x3-x1)*(x2-x1) + (y3-y1)*(y2-y1))/(((x3-x1)**2+(y3-y1)**2)**0.5 * ((x2-x1)**2+(y2-y1)**2)**0.5)
    return math.degrees(math.acos(cosabc))
print(degree(diem))
```

```python
"""Lấy đề từ file ôn tập"""
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    wait = 0
    if wait_minutes > 5:
        block = (wait_minutes - 1) // 5
        wait = block * 2000
    s = 0
    if car_type == 4:
        if km <= 0.8 :
            s = 11000 
        elif 0.8 < km <= 20:
            s = 11000 + (km-0.8)*13000
        else:
            s = 11000 + (20-0.8)*13000 + (km - 20)*10500 
    if car_type == 7:
        if km <= 0.8 :
            s = 12000 
        elif 0.8 < km <= 20:
            s = 12000 + (km-0.8)*14000 
        else:
            s = 12000 + (20-0.8)*14000 + (km - 20)*12000 
    total = s + wait
    if start_hour in [22,23,0,1,2,3,4,5]:
        total = 1.2 * total
    elif start_hour not in [22,23,0,1,2,3,4,5] and km > 30 :
        total = 0.9 * total
    
    if total < 50000 and km > 5:
        total = 50000
    
    return int(total)
car_type, start_hour, km, wait_minutes = map(float, input().split())
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))
```
