W6A10

```python
a = list(map(int,input().split()))
k = int(input())
ans = []
checkans = set()
for i in range(len(a)):
    for j in range(i+1,len(a)):
        if a[i] + a[j] == k:
            b = (a[i],a[j])
            if b not in checkans:
                ans.append(b)
                checkans.add(b)

ans.sort()
print(ans)
```

```python
a = list(map(int,input().split()))
k = int(input())

check = set()
ans = []
for i in a:
    if k - i in check:
        ans.append((k-i, i))
    
    check.add(i)
ans.sort()
print(ans)
```

```python
n = int(input())
a = []
chinh = []
phu = []
for i in range(1,n+1):
    while True:
        try:
            hang = list(map(int, input().split()))
            if len(hang) != n:
                print("try again")
                continue
            break
        except ValueError:
            print("try again")
    a.append(hang)
for j in range(n):
    for k in range(n):
        if n%2 == 0:
            if j == k:
                chinh.append(a[j][k])
            elif j + k == (n - 1):
                phu.append(a[j][k])
        else:
            if j == k: 
                chinh.append(a[j][k])
            if j+k ==(n-1) or j == k == (n//2) :
                phu.append(a[j][k]) 
print(chinh)
print(phu)
```
