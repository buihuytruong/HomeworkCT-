```python
#W4A7-2
n = int(input())
def unt(n):
    while n % 2 == 0:
        a = 2
        n //=2 
    for i in range(3, n , 2):
        while n % i == 0:
            a = i
            n //= i
    if n > 2:
        a = n 
    return a
print(unt(n))
```

```python
#W4A14
def ucln(a,b):
    c =[]
    d = 0
    for i in range(1,a+1):
        if a % i == 0 and b%i == 0:
            c.append(i)
        for k in c:
            if k > d:
                d = k
    return k
a , b = map(int , input().split())
print(ucln(a,b))
    
```
