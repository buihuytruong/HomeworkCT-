### W9A1 Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ thuật toán sắp xếp nào


```python
def sort(a,b):
    i = 0
    j = 0 
    ans = []
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            ans.append(a[i])
            i += 1
        else:
            ans.append(b[j])
            j += 1
    while i < len(a) :
        ans.append(a[i])
        i += 1
    while j < len(b) :
        ans.append(b[j])
        j += 1
    return ans
a = list(map(int,input().split()))
b = list(map(int,input().split()))
print(*sort(a,b))
```

### W9A2 Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên dòng tăng dần và trên cột tăng dần, hãy tìm số lớn thứ k (k <nxm) với thuật toán tốt nh

```python
n , m = map(int,input().split())
k = int(input())
```

```python
def count_less_equal(matrix, mid):
    """Hàm đếm số lượng phần tử <= mid trong thời gian O(n + m)"""
    count = 0
    n = len(matrix)
    m = len(matrix[0])
    
    # Bắt đầu từ góc dưới cùng bên trái của ma trận
    row = n - 1
    col = 0
    
    while row >= 0 and col < m:
        if matrix[row][col] <= mid:
            # Nếu phần tử hiện tại <= mid, thì tất cả các phần tử 
            # phía trên nó trong cùng một cột cũng sẽ <= mid
            count += (row + 1)
            col += 1  # Tiến sang cột bên phải
        else:
            row -= 1  # Lùi lên dòng phía trên
            
    return count

def find_kth_smallest(matrix, k):
    """Hàm tìm phần tử nhỏ thứ k sử dụng Tìm kiếm nhị phân"""
    low = matrix[0][0]
    high = matrix[-1][-1]
    
    ans = low
    while low <= high:
        mid = (low + high) // 2
        if count_less_equal(matrix, mid) >= k:
            ans = mid       # Ghi nhận đáp án tiềm năng
            high = mid - 1  # Thử tìm giá trị nhỏ hơn ở nửa trái
        else:
            low = mid + 1   # Tìm ở nửa phải
            
    return ans

def main():
    import sys
    input_data = sys.stdin.read().splitlines()
    
    if not input_data:
        return

    matrix = []
    for line in input_data:
        line = line.strip()
        if not line:
            continue
        row_elements = list(map(int, line.split()))
        
        # Nếu dòng chỉ chứa đúng 1 số, đó chính là giá trị k
        if len(row_elements) == 1:
            k = row_elements[0]
            # Tiến hành xử lý ma trận hiện tại và in kết quả
            if matrix:
                print(find_kth_smallest(matrix, k))
                matrix = [] # Reset để xử lý test case tiếp theo (nếu có)
        else:
            matrix.append(row_elements)

if __name__ == "__main__":
    main()
```

### W9A3 Cho một số nguyên dương, hãy thực hiện hoán đổi đúng 2 chữ số để được một số nguyên nhỏ nhất có thể 


```python
n = int(input())
s = str(n)
min_n = float('inf')
def ans(s):
    a = s.lstrip('0')
    return a if a else "0"
for i in range(len(s)):
    for j in range(i+1,len(s)):
        slist = list(s)
        slist[i] , slist[j] = slist[j] , slist[i]
        new = int(ans(''.join(slist)))
        if new < min_n:
            min_n = new
print(min_n)
```
