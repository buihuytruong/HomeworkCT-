```python
#W2A3 dung ham def
def tinhtoan(a, b):
    tong = a + b
    hieu = a - b
    tich = a * b
    pnguyen = a // b
    pdu = a % b
    return tong, hieu , tich , pnguyen , pdu 
a , b = map(int, input().split())
ans = tinhtoan(a,b)
for i in range(5):
    print(ans[i])
```
